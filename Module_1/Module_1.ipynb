````python
# %load https://tinyurl.com/pyshell-deco-ex
"""
Exercise:
     Implement the command_dispatch module with class CommandDispatch
     such that the following example program that implements a
     rudimentary command-line shell works.
"""

from command_dispatch import CommandDispatch


shell = CommandDispatch()

@shell.for_command("list")
def list_directory(*args):
    from os import listdir
    if len(args) < 2:
        args += ".",
    for path in args[1:]:
        print("{}:".format(path))
        print("\n".join(listdir(path)))

@shell.for_command("whoami")
def show_user(*args):
    from getpass import getuser
    print(getuser())

@shell.for_command("date")
def print_date(*args):
    from time import ctime
    print(ctime())

@shell.for_command("pwd")
def show_curr_dir(*args):
    import os
    print(os.getcwd())

@shell.for_command("exit")
def exit_shell(*args):
    exit(0)

@shell.for_command("hostname")
def show_hostname(*args):
    from os import uname
    print(uname().nodename)

@shell.invalid
def invalid_command(*args):
    print("Invalid command - ", args[0])


@shell.input
def get_input():
    import rlcompleter
    return input("PyShell> ").split()

if __name__ == '__main__':
    shell.run()

````

### NOTE:
> Always design functions in python as "pure-functions"
> If you need to implement functions with 'side-effects', prefer creating classes and
> implement those side-effects as 'instance methods' within class definitions.

````python
# Pure functions are functions that do not carry states across calls.
# They always produce the same output for the same input.

def square(x):  # pure function
    return x*x

print(square(2))
print(square(3))
````

````python
total = 0

def accumulate(x): # function with side-effects
    global total
    total += x
    return total

print(accumulate(10))
print(accumulate(10))
print(accumulate(10))
````

````python
class Car:
    pass

c1 = Car()
c2 = Car()

c1.make = "Honda" # Setting instance attributes manually.
c2.make = "Toyota"
print(c1.make, c2.make)

def drive_car(c):
    print("Driving a", c.make)

drive_car(c1)
drive_car(c2)
````

````python
Car.drive = drive_car # Assigning a callable object (function) as class attribute -> promotes it to an instance method
Car.drive(c1)
Car.drive(c2)

c1.drive()  # Instance method call
c2.drive()
````

````python
class Car:
    def drive(self): # Class attribute that becomes an instance method
        print("Driving a", self.make)

    def sell():
        print("Sold this car...")


c1 = Car()
c2 = Car()
c1.make = "Honda"
c2.make = "Toyota"
c1.drive() # Instance method call -> Car.drive(c1)
c2.drive()

c1.sell() # Car.sell(c1) -> Error
````

````python
# The role of __init__ method in classes is to initialize instance attributes.
# The __init__ method is called automatically when an instance is created.

class Car:
    def __init__(self, make):
        self.make = make

    def drive(self):
        print("Driving a", self.make)

c1 = Car("Honda")
c2 = Car("Toyota")
c1.drive()
c2.drive()
````

````python
class Car:
    def __init__(self):
        print("__init__ invoked: self =", self)

c = Car()
print(c)
````

````python
class Car:
    def __init__(self):
        print("__init__ invoked: self =", self)
        return "Hello"
    
c = Car()
print(c)
````

````python
class Car:
    def __new__(cls):
        print("__new__ invoked: cls =", cls)
        return 500

c = Car()
print(c)
````
